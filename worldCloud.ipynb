```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('word cloud.csv')
df.head()
```

```python

new_df = df.loc[df.index.repeat(df.number)]
words = new_df.words
text = " ".join(t.replace(' ', '-') for t in words)
text
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 1000, height = 1000, background_color="white", 
                        contour_color='firebrick', stopwords = stopwords, 
                        min_font_size = 12).generate(text)
plt.figure(figsize = (12, 12), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.show() 
wordcloud.to_file('wordcloud.eps')
```

```python
# Generate word cloud from frequencies
d = dict(zip(df.words, df.number))
wordcloud = WordCloud(width = 1000, height = 800, background_color="white", 
                        contour_color='firebrick', stopwords = stopwords, 
                        min_font_size = 12).generate_from_frequencies(d)
plt.figure(figsize = (12, 8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show() 
wordcloud.to_file('wordcloud.eps')
```
